```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons,make_gaussian_quantiles
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc

# Generate sample data
X, _ = make_gaussian_quantiles(n_samples=100,  random_state=0)

dendro=shc.dendrogram(shc.linkage(X))
```

```python
dbscan = AgglomerativeClustering(n_clusters=2)
y_dbscan = dbscan.fit_predict(X)

# Plotting the clusters
plt.figure(figsize=(10, 6))

# Plotting the points with different colors for different clusters
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', s=50)

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
