```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import  DecisionTreeClassifier
from sklearn.svm import SVC
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
can=load_iris()
df=pd.DataFrame(can.data,columns=can.feature_names)
df['target']=can.target
df.sample(3)
```

```python
X=can.data[:,0:2]
y=can.target
```

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
X_train ,X_test,y_train, y_test =train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=BaggingClassifier(estimator=RandomForestClassifier(max_leaf_nodes=6,min_samples_leaf=5),n_estimators=10,n_jobs=-1,verbose=True)

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```

```python
#classic bagging -> bootstrap=True

model=BaggingClassifier(estimator=RandomForestClassifier(max_leaf_nodes=6,min_samples_leaf=5),bootstrap=True,n_estimators=10,n_jobs=-1,verbose=True)

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```

```python
#pasting -> bootstrap=False

model=BaggingClassifier(estimator=RandomForestClassifier(max_leaf_nodes=6,min_samples_leaf=5),bootstrap=False,n_estimators=10,n_jobs=-1,verbose=True)

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```

```python
#random subspaces -> bootstrap=False,bootstrap_features=True

model=BaggingClassifier(estimator=RandomForestClassifier(max_leaf_nodes=6,min_samples_leaf=5),bootstrap=False,bootstrap_features=True,n_estimators=10,n_jobs=-1,verbose=True)

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```

```python
#random patches -> bootstrap=False,bootstrap_features=True

model=BaggingClassifier(estimator=RandomForestClassifier(max_leaf_nodes=6,min_samples_leaf=5),bootstrap=True,bootstrap_features=True,n_estimators=10,n_jobs=-1,verbose=True)

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```
