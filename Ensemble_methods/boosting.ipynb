```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
can=load_iris()
df=pd.DataFrame(can.data,columns=can.feature_names)
df['target']=can.target
df.sample(3)
```

```python
X=can.data[:,0:2]
y=can.target
```

AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
X_train ,X_test,y_train, y_test =train_test_split(X,y,test_size=0.3,random_state=42)

model=AdaBoostClassifier(estimator=RandomForestClassifier(max_depth=1),n_estimators=60,learning_rate=0.1)

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```

gradient boost

```python
from sklearn.ensemble import GradientBoostingClassifier
model1=GradientBoostingClassifier(n_estimators=60,learning_rate=0.05,max_depth=1)

model1.fit(X_train,y_train)
y_pred=model1.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```

XGBoost

```python
import xgboost as xgb

model2=xgb.XGBClassifier(objective='multi:softmax',num_class=3,n_estimators=25,learning_rate=0.1)

model2.fit(X_train,y_train)
y_pred=model2.predict(X_test)
print(accuracy_score(y_test,y_pred))

plot_decision_regions(X_train,y_train,clf=model)
```
