```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df=df[['sex','age','class','embark_town','alive','survived']]
df
```

```python
# SimpleImputer for fill missing values
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test ,y_train ,y_test=train_test_split(df.iloc[:,0:5],df.iloc[:,-1],test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[
    ('ct1',SimpleImputer(),['age']),
    ('ct2',OrdinalEncoder(categories=[['First','Second','Third']]),['class']),
    ('cl3',OneHotEncoder(sparse_output=False,drop='first'),['embark_town'])
    ],remainder='passthrough')
```

```python
X_train_transformed=ct.fit_transform(X_train)
X_test_transformed=ct.transform(X_test)
```

```python
column_names = ['age', 'class', 'embark_town_Cherbourg', 'embark_town_Queenstown', 'embark_town_Southampton', 'sex', 'alive']
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=column_names)
X_train_transformed_df
```
