```python
import seaborn  as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
df=df.dropna()
```

```python
df=df[['sex','class','alone','alive']]
df
```

using pandas get dummies

```python
df_dummies=pd.get_dummies(df,columns=['class','alone'])
for col in df_dummies.select_dtypes(include=['bool']).columns:
    df_dummies[col] = df_dummies[col].astype(int)

df_dummies
```

using sklearn

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test=train_test_split(df.iloc[:,0:3],df.iloc[:,-1],test_size=0.3)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=int)

X_train_trans=ohe.fit_transform(X_train)
X_test_trans=ohe.transform(X_test)
X_test_trans
```
