```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df=df.dropna()
df
```

```python
df['embarked'].value_counts().sum
```

```python
df=df[['class','embarked','alive']]
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train,y_test=train_test_split(X,y,test_size=0.3)

```

```python
from sklearn.preprocessing import OrdinalEncoder

ore=OrdinalEncoder(categories=[['First','Second','Third'],['S','C','Q']],dtype=np.int32)

X_train_transformed=ore.fit_transform(X_train)
X_test_transformed=ore.transform(X_test)
```

```python

df_new1=pd.DataFrame(X_train_transformed)
df_new2=pd.DataFrame(X_test_transformed)

df_new=pd.concat((df_new1,df_new2),axis=0)
df_new.sample(5)
```
