```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('penguins')
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X=df.iloc[:,2:6]
y=df.iloc[:,-1]

X_train ,X_test, y_train,y_test=train_test_split(X,y,test_size=0.3)

sd=StandardScaler()

X_train_scaled=sd.fit_transform(X_train)
X_test_scaled=sd.transform(X_test)


df_train_new=pd.DataFrame(X_train_scaled,columns=[X_train.columns])
df_test_new=pd.DataFrame(X_test_scaled,columns=[X_test.columns])
df_new=pd.concat([df_train_new,df_test_new],axis=0)
df_new
```
