```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df=df[['class','sex','embarked','alone','alive']]
df
```

```python
df.isnull().sum()
```

```python
df['embarked'].value_counts().sum
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.3)
```

```python
X_train
```

```python
tfm=ColumnTransformer(transformers=[
    ('embarked_fill',SimpleImputer(strategy='most_frequent'),[2]),

    ],remainder='passthrough')
```

```python
yak=tfm.fit_transform(X_train)
dt=pd.DataFrame(yak)
dt[0].isnull().sum()
```

```python
transform1=ColumnTransformer(transformers=[
    ('embarked_fill',SimpleImputer(strategy='most_frequent'),[2])
     ],remainder='passthrough')
```

```python

transform2=ColumnTransformer(transformers=[
    ('class_ohe',OneHotEncoder(categories=[['First','Second','Third']]),[0]),
    ('embarked_ohe', OneHotEncoder(categories=[['S', 'C', 'Q']]),[2])
    ],remainder='passthrough')
```

```python
transform3=ColumnTransformer(transformers=[
    ('sex_oe',OrdinalEncoder(handle_unknown='ignore'),[1]),
    ('alone_oe',OrdinalEncoder(handle_unknown='ignore'),[3]),
    ('alive_oe',OrdinalEncoder(handle_unknown='ignore'),[4])       
    ],remainder='passthrough')
```

```python
transform4=DecisionTreeClassifier()
```

creating pipeline

```python
pipe=Pipeline([
    ('transform1',transform1),
    ('transform2',transform2),
    ('transform3',transform3),
    ('transform4',transform4),
])
```
